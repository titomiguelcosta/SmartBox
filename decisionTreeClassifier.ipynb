```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
X = data.drop(columns=["label"]) # use alll columns but label
X = data.iloc[:,[0, 2, 4, 6]] # use only certain columns
X.values
```

```python
y = data["label"]
y.values
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X.values, y.values)
```

```python
X.info()
```

```python
predictions = model.predict([ [0.22911, 106.5677, 1450.122, 2022.344], [0.23911, 116.5677, 1750.122, 3022.344] ])
predictions
```

```python
# train/test data and determine accuracy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
accuracy_score(y_test, predictions)
```

```python
import joblib
```

```python
joblib.dump(model, "model.joblib")
```
